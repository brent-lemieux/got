```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data/got_deaths.csv')
```

```python
df.head().T
```

```python
df.info()
```

```python
by_ep = df.groupby('death_episode', as_index=False)['name'].count()
by_ep
```

```python
by_seas = df.groupby(['death_season', 'death_episode'], as_index=False)['name'].count()
by_seas.columns = ['death_season', 'death_episode', 'death_count']
by_seas
```

```python
plt.figure(figsize=(12,6))

for season in np.unique(by_seas['death_season']):
    seas_df = by_seas.loc[by_seas['death_season']==season]
    plt.plot(seas_df['death_episode'], seas_df['death_count'], label=season)
    plt.legend()
    
plt.xticks(list(range(11)))
plt.title('Death Episode by Season')
plt.savefig('images/death_episode_by_season.png')
plt.show()
```
